**Objetivo:**  
Seleccionar dos fotografías de una misma persona y una tercera fotografía de una persona diferente. Calcular las distancias euclidianas entre las imágenes y realizar una comparación de los resultados

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import distance

# Cargar imágenes
img1 = Image.open("JamesSeleccion.webp").convert("L")  # Convertir a escala de grises
img2 = Image.open("JamesLeon.jfif").convert("L")
img3 = Image.open("JhonArias.jpg").convert("L")

# Mostrar imágenes
fig, axs = plt.subplots(1, 3, figsize=(12,4))
axs[0].imshow(img1, cmap="gray"); axs[0].set_title("Imagen 1")
axs[1].imshow(img2, cmap="gray"); axs[1].set_title("Imagen 2")
axs[2].imshow(img3, cmap="gray"); axs[2].set_title("Imagen 3")
plt.show()
```

```python
# Redimensionar imágenes a 128x128 para comparación
size = (128,128)
img1 = img1.resize(size)
img2 = img2.resize(size)
img3 = img3.resize(size)

#Mostrar las imágenes tras el proceso de redimensionamiento
fig, axs = plt.subplots(1, 3, figsize=(12,4))
axs[0].imshow(img1, cmap="gray"); axs[0].set_title("Imagen 1")
axs[1].imshow(img2, cmap="gray"); axs[1].set_title("Imagen 2")
axs[2].imshow(img3, cmap="gray"); axs[2].set_title("Imagen 3")
plt.show()

# Convertir a numpy arrays
vec1 = np.array(img1).flatten()
vec2 = np.array(img2).flatten()
vec3 = np.array(img3).flatten()
```

**Distancia Euclideana:**  
La fórmula de la distancia euclidiana entre dos vectores $x$ y $y$ es:  

$$
d(x, y) = \sqrt{\sum_{i} (x_i - y_i)^2}
$$


```python
#Calcular Distancias
d12 = distance.euclidean(vec1, vec2)
d13 = distance.euclidean(vec1, vec3)
d23 = distance.euclidean(vec2, vec3)

print("Distancia entre Imagen 1 y 2:", round(d12,3))
print("Distancia entre Imagen 1 y 3:", round(d13,3))
print("Distancia entre Imagen 2 y 3:", round(d23,3))
```

```python
#Determinar cuáles son las imágenes más parecidas considerando la distancia más pequeña
distancias = {"1-2": d12, "1-3": d13, "2-3": d23}
mas_parecidas = min(distancias, key=distancias.get)

print("Las imágenes más parecidas son:", mas_parecidas,
      "con distancia =", round(distancias[mas_parecidas],3))
```

**Conclusion:**  
Se logró mostrar, mediante el cálculo de la distancia euclidiana entre las imágenes, que las más parecidas son la Imagen 1(James rodriguez en la seleccion) y la Imagen 2(James Rodriguez en Leon de Mexico). Este resultado concuerda con la percepción visual y la lógica del cerebro humano
