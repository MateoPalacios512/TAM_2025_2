```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Proyecci√≥n anal√≠tica (m√≠nimos cuadrados)
# -------------------------------
def proyeccion_analitica(a: np.ndarray, b: np.ndarray) -> float:
    """
    Retorna el escalar x tal que x * a es la proyecci√≥n de b sobre a.
    Utiliza la f√≥rmula anal√≠tica de m√≠nimos cuadrados:
    x = (a^T b) / (a^T a)
    """
    x = np.dot(a, b) / np.dot(a, a)
    return x

# -------------------------------
# 2. Proyecci√≥n por gradiente descendente
# -------------------------------
def gradiente_descendente(a: np.ndarray, b: np.ndarray, lr=0.1, epochs=100) -> float:
    """
    Resuelve el problema de m√≠nimos cuadrados: encontrar x que minimice ||x*a - b||^2
    usando gradiente descendente.
    """
    x = 0.0  # Inicializaci√≥n
    for i in range(epochs):
        # Gradiente de la funci√≥n de p√©rdida L(x) = ||x*a - b||^2
        grad = 2 * np.dot(a, x * a - b)
        x = x - lr * grad
    return x

# -------------------------------
# 3. Graficar los vectores
# -------------------------------
def graficar_proyeccion(a, b, x_analitico, x_iterativo):
    """
    Muestra los vectores originales y sus proyecciones en un gr√°fico.
    """
    origen = np.array([0, 0])
    proj_analitica = x_analitico * a
    proj_iterativa = x_iterativo * a

    plt.figure(figsize=(8, 8))
    plt.quiver(*origen, *a, color='blue', angles='xy', scale_units='xy', scale=1, label='Vector a (base)')
    plt.quiver(*origen, *b, color='green', angles='xy', scale_units='xy', scale=1, label='Vector b (original)')
    plt.quiver(*origen, *proj_analitica, color='red', angles='xy', scale_units='xy', scale=1, label='Proyecci√≥n Anal√≠tica')
    plt.quiver(*origen, *proj_iterativa, color='orange', angles='xy', scale_units='xy', scale=1, label='Proyecci√≥n Iterativa')

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.grid()
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("Proyecci√≥n de un vector sobre otro (Anal√≠tica vs Iterativa)")
    plt.show()

# -------------------------------
# 4. C√≥digo principal
# -------------------------------
def main():
    # Definimos los vectores en R^2
    a = np.array([2, 1])
    b = np.array([3, 4])

    # Soluci√≥n anal√≠tica
    x_analitico = proyeccion_analitica(a, b)
    print(f"Escalar anal√≠tico x: {x_analitico:.4f}")

    # Soluci√≥n por gradiente descendente
    x_iterativo = gradiente_descendente(a, b, lr=0.05, epochs=100)
    print(f"Escalar iterativo x: {x_iterativo:.4f}")

    # Visualizaci√≥n
    graficar_proyeccion(a, b, x_analitico, x_iterativo)

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Proyecci√≥n anal√≠tica (m√≠nimos cuadrados)
# -------------------------------
def proyeccion_analitica(a: np.ndarray, b: np.ndarray) -> float:
    x = np.dot(a, b) / np.dot(a, a)
    return x

# -------------------------------
# 2. Proyecci√≥n por gradiente descendente
# -------------------------------
def gradiente_descendente(a: np.ndarray, b: np.ndarray, lr=0.1, epochs=100) -> float:
    x = 0.0  # Inicializaci√≥n
    for i in range(epochs):
        grad = 2 * np.dot(a, x * a - b)
        x = x - lr * grad
    return x

# -------------------------------
# 3. Graficar proyecci√≥n anal√≠tica
# -------------------------------
def graficar_proyeccion_analitica(a, b, x_analitico):
    origen = np.array([0, 0])
    proj = x_analitico * a

    plt.figure(figsize=(6, 6))
    plt.quiver(*origen, *a, color='blue', scale_units='xy', angles='xy', scale=1, label='Vector a')
    plt.quiver(*origen, *b, color='green', scale_units='xy', angles='xy', scale=1, label='Vector b')
    plt.quiver(*origen, *proj, color='red', scale_units='xy', angles='xy', scale=1, label='Proyecci√≥n Anal√≠tica')

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("Proyecci√≥n Anal√≠tica")
    plt.show()

# -------------------------------
# 4. Graficar proyecci√≥n iterativa
# -------------------------------
def graficar_proyeccion_iterativa(a, b, x_iterativo):
    origen = np.array([0, 0])
    proj = x_iterativo * a

    plt.figure(figsize=(6, 6))
    plt.quiver(*origen, *a, color='blue', scale_units='xy', angles='xy', scale=1, label='Vector a')
    plt.quiver(*origen, *b, color='green', scale_units='xy', angles='xy', scale=1, label='Vector b')
    plt.quiver(*origen, *proj, color='orange', scale_units='xy', angles='xy', scale=1, label='Proyecci√≥n Iterativa')

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("Proyecci√≥n Iterativa (Gradiente Descendente)")
    plt.show()

# -------------------------------
# 5. C√≥digo principal
# -------------------------------
def main():
    a = np.array([2, 1])
    b = np.array([3, 4])

    x_analitico = proyeccion_analitica(a, b)
    x_iterativo = gradiente_descendente(a, b, lr=0.05, epochs=100)

    print(f"Escalar anal√≠tico x: {x_analitico:.4f}")
    print(f"Escalar iterativo x: {x_iterativo:.4f}")

    graficar_proyeccion_analitica(a, b, x_analitico)
    graficar_proyeccion_iterativa(a, b, x_iterativo)

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Funci√≥n de p√©rdida
# -------------------------------
def loss(a, b, x):
    """
    L(x) = ||x * a - b||^2
    """
    return np.linalg.norm(x * a - b) ** 2

# -------------------------------
# 2. Gradiente descendente con tracking
# -------------------------------
def gradiente_descendente_tracking(a, b, lr=0.1, max_epochs=1000, tol=1e-6):
    """
    Descenso por gradiente que guarda la trayectoria de x y L(x)
    para visualizar la convergencia.
    """
    x = 0.0
    history_x = []
    history_loss = []

    for i in range(max_epochs):
        grad = 2 * np.dot(a, x * a - b)
        current_loss = loss(a, b, x)

        # Guardamos el estado actual
        history_x.append(x)
        history_loss.append(current_loss)

        # Criterio de parada: gradiente muy peque√±o
        if abs(grad) < tol:
            print(f"Gradiente peque√±o en iteraci√≥n {i}, deteniendo entrenamiento.")
            break

        # Actualizaci√≥n
        x = x - lr * grad

    return x, history_x, history_loss

# -------------------------------
# 3. Visualizar par√°bola y trayectoria
# -------------------------------
def graficar_parabola(a, b, history_x, history_loss, x_analitico):
    x_vals = np.linspace(min(history_x) - 1, max(history_x) + 1, 300)
    y_vals = [loss(a, b, x) for x in x_vals]

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label='Funci√≥n de p√©rdida $L(x)$', color='blue')
    plt.plot(history_x, history_loss, 'o-', color='red', label='Descenso gradiente')
    plt.axvline(x_analitico, color='green', linestyle='--', label='M√≠nimo anal√≠tico')
    plt.xlabel('$x$')
    plt.ylabel('$L(x)$')
    plt.title('Descenso por gradiente sobre la par√°bola de p√©rdida')
    plt.legend()
    plt.grid(True)
    plt.show()

# -------------------------------
# 4. Proyecci√≥n anal√≠tica
# -------------------------------
def proyeccion_analitica(a: np.ndarray, b: np.ndarray) -> float:
    return np.dot(a, b) / np.dot(a, a)

# -------------------------------
# 5. Main
# -------------------------------
def main():
    a = np.array([2, 1])
    b = np.array([3, 4])

    x_analitico = proyeccion_analitica(a, b)
    print(f"Soluci√≥n anal√≠tica: x = {x_analitico:.4f}")

    x_iterativo, history_x, history_loss = gradiente_descendente_tracking(a, b, lr=0.05, max_epochs=200, tol=1e-6)
    print(f"Soluci√≥n iterativa: x = {x_iterativo:.4f}")
    print(f"√öltima p√©rdida: L(x) = {history_loss[-1]:.6f}")

    graficar_parabola(a, b, history_x, history_loss, x_analitico)

if __name__ == "__main__":
    main()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------
# Funciones de comparaci√≥n
# -------------------------------
def proyeccion_escalares(a, b):
    """Devuelve el escalar de la proyecci√≥n de b sobre a (protege divisi√≥n por cero)."""
    denom = np.dot(a, a)
    return (np.dot(a, b) / denom) if denom != 0 else 0.0

def similitud_coseno(a, b):
    """Devuelve la similitud coseno entre dos vectores (protege normas cero)."""
    na = np.linalg.norm(a)
    nb = np.linalg.norm(b)
    return (np.dot(a, b) / (na * nb)) if (na > 0 and nb > 0) else 0.0

# -------------------------------
# Utilidades de I/O y visualizaci√≥n
# -------------------------------
def cargar_imagenes(filenames):
    imgs = []
    for fn in filenames:
        imgs.append(Image.open(fn).convert("L"))
    return imgs

def mostrar_imagenes(imlist, titles=None, figsize=(12,4)):
    n = len(imlist)
    fig, axs = plt.subplots(1, n, figsize=figsize)
    for i, img in enumerate(imlist):
        ax = axs[i] if n > 1 else axs
        ax.imshow(img, cmap="gray")
        ax.set_title(titles[i] if titles else f"Imagen {i+1}")
        ax.axis("off")
    plt.tight_layout()
    plt.show()

# -------------------------------
# Programa principal
# -------------------------------
def main():
    # Cambia los nombres de archivo aqu√≠ si quieres
    filenames = ["JamesSeleccion.webp", "JamesLeon.jfif", "JhonArias.jpg"]

    # 1) Cargar y mostrar originales
    imgs = cargar_imagenes(filenames)
    mostrar_imagenes(imgs, titles=[f"Imagen {i+1}\n{fn}" for i, fn in enumerate(filenames)])

    # 2) Redimensionar y mostrar
    size = (128, 128)
    imgs_resized = [im.resize(size) for im in imgs]
    mostrar_imagenes(imgs_resized, titles=[f"Imagen {i+1} (resized)" for i in range(len(imgs_resized))])

    # 3) A vectores (float64 para estabilidad)
    vecs = [np.array(im).flatten().astype(np.float64) for im in imgs_resized]

    # 4) Similitud coseno (mismo orden: 1-2, 1-3, 2-3)
    pairs = [(0, 1), (0, 2), (1, 2)]
    keys = ["1-2", "1-3", "2-3"]
    cos_vals = [similitud_coseno(vecs[i], vecs[j]) for i, j in pairs]
    cos_arr = np.array(cos_vals)

    # 5) Escalares de proyecci√≥n (mismo orden: 2 sobre 1, 3 sobre 1, 3 sobre 2)
    proj_vals = [
        proyeccion_escalares(vecs[0], vecs[1]),  # 2 sobre 1
        proyeccion_escalares(vecs[0], vecs[2]),  # 3 sobre 1
        proyeccion_escalares(vecs[1], vecs[2])   # 3 sobre 2
    ]

    # Impresi√≥n de resultados
    for k, v in zip(keys, cos_arr):
        print(f"Similitud coseno {k}: {v:.4f}")
    best_idx = int(np.argmax(cos_arr))
    print(f"\nLas im√°genes m√°s parecidas son {keys[best_idx]} con similitud {cos_arr[best_idx]:.4f}\n")

    print("Escalares de proyecci√≥n:")
    for label, val in zip(["2 sobre 1", "3 sobre 1", "3 sobre 2"], proj_vals):
        print(f"{label}: {val:.4f}")

    # 6) Gr√°ficas: resaltar la barra m√°s parecida
    colors = ["skyblue"] * len(keys)
    colors[best_idx] = "orange"

    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Similitud coseno (resaltada)
    bars = axs[0].bar(keys, cos_arr, color=colors)
    axs[0].set_ylim(0, 1)
    axs[0].set_ylabel("Similitud coseno")
    axs[0].set_title("Similitud coseno entre im√°genes")
    for idx, bar in enumerate(bars):
        h = bar.get_height()
        axs[0].text(bar.get_x() + bar.get_width() / 2, h + 0.02, f"{h:.4f}", ha="center", va="bottom")
        if idx == best_idx:
            axs[0].text(bar.get_x() + bar.get_width() / 2, h + 0.08, "‚Üê M√°s parecidas", ha="center", va="bottom", fontsize=10, color="orange")

    # Escalares de proyecci√≥n
    bars2 = axs[1].bar(["2 sobre 1", "3 sobre 1", "3 sobre 2"], proj_vals, color="lightgreen")
    axs[1].set_title("Escalares de proyecci√≥n")
    for bar in bars2:
        h = bar.get_height()
        # ajustar posici√≥n del texto seg√∫n signo del valor
        if h >= 0:
            axs[1].text(bar.get_x() + bar.get_width() / 2, h + 0.02, f"{h:.4f}", ha="center", va="bottom")
        else:
            axs[1].text(bar.get_x() + bar.get_width() / 2, h - 0.02, f"{h:.4f}", ha="center", va="top")

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
